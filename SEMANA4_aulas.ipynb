# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

```python
#fazendo a média de valores total
dados['Valor'].mean()
```

```python
#agora, a media agrupando em "tipos"
dados.groupby("Tipo").mean(numeric_only=True) #indicando qual coluna agrupar e depois a operação de media
#indicamos tambem, que devemos calcular apenas valores numericos
```

```python
#Para melhor visualização
dados.groupby("Tipo")['Valor'].mean() #coluna só de valors - não necessita especificar
```

```python
#O valor gerado acima é uma Series, para visualizar melhor, transformaremos em DataFrame
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor') #sort_values os ordenara em ordem crescente
```

```python
#criando grafico; salvar DataFrame como variavel
preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
#kind=barh p grafico de barras horizontais, figsize é o tamanho do grafico a plotar e color a cor
preco_tipo.plot(kind='barh', figsize=(12, 10), color ='pink')
```

##Parâmetros por trás do groupby
### **by**: mais comum; especifica a(s) coluna(s) a agrupar; como argumento, passamos o nome da coluna ou lista de nomes de colunas.
### **axis**: eixo a ser agrupado, o padrao é zero(linhas), para agrupar colunas usamos eixo 1.
###**sort**: booleano que indica se os grupos devem ser ordenados pelo valor da coluna de agrupamento.
###**dropna**: controlar se os valores NaN sera exluidos ou nao, o padrão é True.

```python
import pandas as pd

df = pd.DataFrame({
   'Animal': ['Cachorro', 'Gato', 'Elefante', 'Cachorro', 'Gato', 'Elefante'],
   'Cor': ['Preto', 'Branco', 'Cinza', 'Marrom', 'Preto', 'Marrom'],
   'Quantidade': [2, 3, 1, 4, 2, 2]
})
df
```
